```python
import csv
from bs4 import BeautifulSoup

# Open the file and read the HTML content
with open("page_content_y15.txt", "r", encoding="utf-8") as file:
    html_content = file.read()

# Parse the HTML content with BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")

file_name = "output_y15.csv"
websiteLink = "https://www.carousell.sg"
keys = ['Username', 'Name', 'Before Price', 'After Price', 'Link']
data = {key: [] for key in keys}
links = soup.find_all("a")
# print(links)
for link in links:
    href = link.get("href")
    
    if href[:3] == "/u/":
        userName = href[3:].split("/")[0]
        data['Username'].append(userName)
    
    if href[:3] == "/p/":  # Ensure there is an href attribute
        try:
            productName, price = link.get_text().split("S$")
            # print("Name: ", productName)
            # print("Price: S$", price)
            # print("link: ", href)
            data['Name'].append(productName)
            data['Before Price'].append("Not Slashed")
            data['After Price'].append(price)
            data['Link'].append(websiteLink + href)
        except ValueError:
            productName, price, beforePrice = link.get_text().split("S$")
            # print("Name: ", productName)
            # print("Price Before: ", beforePrice)
            # print("Price Discounted: S$", price)
            # print("link: ", href)
            data['Name'].append(productName)
            data['Before Price'].append(beforePrice)
            data['After Price'].append(price)
            data['Link'].append(websiteLink + href)
    file_name = 'output.csv'
# print(data)
# Open the file in append mode and create it if it doesn't exist
with open(file_name, 'a', newline='') as file:
    writer = csv.writer(file)
    
    # If the file is empty, write the headers first
    if file.tell() == 0:  # Check if the file is empty
        writer.writerow(data.keys())  # Write dictionary keys as headers
    
    # Write the rows (transpose the dictionary values to rows)
    rows = zip(*data.values())
    writer.writerows(rows)
```

```python
# Run this in jupyter notebook environment
import pandas as pd

# Path to your CSV file
csv_file = 'krr.csv'

# Load the CSV file into a Pandas DataFrame
df = pd.read_csv(csv_file)

# Ensure 'Date' is treated as a proper datetime object (optional, based on your data)
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# Create a new DataFrame where each link appears only once across different dates
filtered_df = df.drop_duplicates(subset='Link', keep='first')

filtered_df.to_csv('filtered_links.csv', index=False)
```

```python
import pandas as pd

# Load the CSV into a DataFrame
csv_file = 'krr.csv'  # Replace with your CSV file path
df = pd.read_csv(csv_file)

# Extract the unique identifier from the Link column
df['Unique_ID'] = df['Link'].str.extract(r'/p/([^/?]+)').drop_duplicates()

# Drop duplicates based on 'Unique_ID'
filtered_df = df.drop_duplicates(subset='Unique_ID', keep='first')

# Print the result
print(filtered_df[['Date', 'Link']])  # Display the original columns

# Save the filtered data to a new CSV
filtered_df.to_csv('filtered_links.csv', index=False)
```
